# 生成诈骗案例知识图谱

该笔记本演示了通过诈骗案例的文本（判决书）数据集，借助大语言模型，生成知识图谱导入到 Neo4j 数据库中。

**主要步骤**
1. 获得诈骗案例数据集，该数据集为各种诈骗案的判决书，包含诈骗案例的文本数据，每个案例包含被告人、被害人、诈骗金额、诈骗手段等信息。
2. 定义一个提示词模板，要求提示词以固定的数据模式 (Schema) 提取知识图谱的实体与关系，将数据集中的文本数据输入到大语言模型中，最终得到可以导入到 Neo4j 数据库的 Cypher 代码。
3. 使用 Cypher 代码导入知识图谱数据到 Neo4j 数据库。

## 使用大语言模型生成 Cypher 代码

### 读入数据集

````python
# 遍历 txt 文件夹及其子文件夹中的所有文本文件
import os
import chardet

txt_folder = "dataset" # 数据集 txt 文件夹路径
texts: dict[str, str] = {}
for root, dirs, files in os.walk(txt_folder):
    for txt_name in files:
        if txt_name.endswith(".txt"):
            file_path = os.path.join(root, txt_name)
            # print(f"正在读取文件：{file_path}")
            with open(file_path, "rb") as file:
                rawdata = file.read()
                encoding = chardet.detect(rawdata)["encoding"] # 检测文件编码
                with open(file_path, "r", encoding=encoding) as file:
                    text = file.read()
                    texts[txt_name] = text
        
````

### 定义提示词模板

提示词模板非常重要，它定义了如何从文本中提取实体和关系，让大语言模型能够理解文本数据，并能按照指定的数据模式生成 Cypher 代码。

````python
# 定义 prompt 模板
prompt_template = """
我正在完成关于反诈骗的 Neo4j 知识图谱构建。请根据我以下给出的案件判决书，以及知识图谱 Schema，写出合并进 Neo4j 的 Cypher 代码。
**要求**
- Cypher 代码中，所有实体与关系按照给定的 Schema 定义，不要翻译中英文。
- 请以 MERGE 命令合并，请只输出代码，不要输出其他内容。
- 设置属性时，不要使用 SET 命令设置，而是在 MERGE 命令中直接设置。如：
  ```cypher
  MERGE (location1:地点 {{name: "浙江省金华市", province: "浙江省", city: "金华市"}})
  ```
- 只有标注为可缺失的属性或实体才可以缺失。
- 代码中不需要注释
- 代码中不需要有 Markdown 格式（如标注 ```cypher```），只需要 Cypher 语句即可。
- 可以生成多个语句，每个语句后不需要有分号
- 每个关系都需要为关系定义变量名（如 `relate` ）：
  ```cypher
  MERGE (case)-[relate:刑事判决 {{type: "有期徒刑", 刑期: duration({{years:6}}), 罚金: 60000}}]->(person1)
  ```
  像这样，`relate` 就是关系的变量名。

## 案件判决书
{text}

## **反诈骗知识图谱数据模式**

### **节点类型 (Node Type)**

#### 1. **案件**
- **属性**：
  - `name`：案件的名称，字符串类型。
  - `type`：案件类型，分类标签，枚举类型，取值范围为：`刑事`、`民事`、`行政`、`执行`、`其他`。
  - `description`：案件的简单描述，文本类型。
  - `date_of_incident`：案发时间，Date 类型，如 `date("2020-01-01")`。（不可缺失）
  - `date_of_judgment`：审判时间，Date 类型，如 `date("2020-01-01")`。（可缺失）

#### 2. **人物**（尽量不要缺失）
- **属性**：
  - `name`：人物的全名，字符串类型。
  - `gender`：枚举类型，取值范围为：`男`、`女`、`未知`。（可缺失）
  - `birthday`：Date 类型。（可缺失）
  - `career`：人物的职业，字符串类型。（可缺失）
  - `education`：人物的学历或教育经历，字符串类型。（可缺失）
  - `description`：关于人物的详细描述，文本类型。

#### 3. **机构**（可缺失）
- **属性**：
  - `name`：机构的全称，字符串类型。
  - `type`：分类标签，字符串类型，可以是：`政府机构`、`企业`、`事业单位`、`社会团体`、 等（可缺失）
  - `description`：关于机构的详细描述，文本类型。

#### 4. **地点**（尽量不要缺失）
- **属性**：
  - `name`：地点的详细地址，字符串类型。
  - `province`：省/自治区/直辖市名称，不需要行政区划名，字符串类型（如“广东”，不要写成“广东省”）。
  - `city`：地级市名称，不需要行政区划名，字符串类型（如“广州”，不要写成“广州市”）。（可缺失）
  - `district`：区/县名称，不需要行政区划名，字符串类型（如“天河”，不要写成“天河区”）。（可缺失）

#### 5. **工具**（尽量不要缺失）
- **属性**：
  - `name`：工具的名称，字符串类型。
  - `type`：分类标签，枚举类型，取值范围为：`社交软件`、`银行卡`、`手机`、`短信`、`其他`。
  - `usage`：工具在诈骗中的用途，字符串类型（如“冒充好友”、“冒充公检法”）。

#### 6. **诈骗类型**（不能缺失！！）
- **属性**：
  - `name`：诈骗类型的名称，字符串类型。
  - `subtype`：诈骗的具体手法，字符串类型（如“冒充公检法”、“虚假投资”）。

#### 7. **实体资产**
- **属性**：
  - `name`：资产的名称，字符串类型。
  - `type`：分类标签，字符串类型，可以为：`房产`、`车辆`、`股票`、`钱财`、等
  - `amount`：资产的数量，数值类型。
  - `unit`：资产的计量单位，字符串类型（如“元”、“平方米”、“套”）。

#### 8. **罪名**（可缺失）
- **属性**：
  - `name`：罪名的全称，字符串类型。

#### 9. **法律法规**（可缺失）
- **属性**：
  - `name`：法律法规的名称，字符串类型。
  - `type`：分类标签，字符串类型，可以为：`刑法`、`民法`、`行政法`、`商法`、`其他`。
  - `description`：法律法规的具体条款或内容，文本类型。

---

### **关系类型 (Relationship Type)**

#### 1. **涉及被害人**
- **定义**：案件与被害人之间的关系。
- **方向**：`案件 -> 人物` 或 `案件 -> 机构`。
- **属性**：
  - 无。

#### 2. **涉及嫌疑人**
- **定义**：案件与嫌疑人或被告人之间的关系。
- **方向**：`案件 -> 人物` 或 `案件 -> 机构`。
- **属性**：
  - 无。

#### 3. **属于组织**（可缺失）
- **定义**：人物隶属于某个机构的关系。
- **方向**：`人物 -> 机构`。
- **属性**：
  - `career`：人物在机构中的职位，字符串类型。

#### 4. **所在地**（可缺失）
- **定义**：人物或机构所在的地理位置。
- **方向**：`人物 -> 地点` 或 `机构 -> 地点`。
- **属性**：
  - 无。

#### 5. **案发地点**（可缺失）
- **定义**：案件发生的位置。
- **方向**：`案件 -> 地点`。
- **属性**：
  - 无。

#### 8. **触犯法律法规**（可缺失）
- **定义**：人物或机构违反的法律法规。
- **方向**：`人物 -> 法律法规` 或 `机构 -> 法律法规`。
- **属性**：
  - 无。

#### 9. **诈骗类型**（不可缺失！！）
- **定义**：案件所属的诈骗类型。
- **方向**：`案件 -> 诈骗类型`。
- **属性**：
  - 无。

#### 11. **涉案工具**
- **定义**：案件中使用的工具。
- **方向**：`案件 -> 工具`。
- **属性**：
  - 无。

#### 12. **人物关系**（可缺失）
- **定义**：人物之间的关联。
- **方向**：`人物 -> 人物`。
- **属性**：
  - `type`：分类标签，字符串类型，可以为：`家人`、`朋友`、`同伙`、`其他` 等。（不可缺失）

#### 13. **涉案资产**（不可缺失！！）
- **定义**：案件中涉及的资产。
- **方向**：`案件 -> 实体资产`。
- **属性**：
  - 无。

#### 14. **罪名**（不可缺失！！）
- **定义**：人物或机构被指控的罪名。
- **方向**：`人物 -> 罪名` 或 `机构 -> 罪名`。
- **属性**：
  - 无。

#### 15. **刑事判决**（可缺失）
- **定义**：案件对人物或机构的判决结果。
- **方向**：`案件 -> 人物` 或 `案件 -> 机构`。
- **属性**：
  - `type`：分类标签，字符串类型，可以为：`缓刑`、`有期徒刑`、`无期徒刑`、`死刑`、`罚款`。（不可缺失）
  - `刑期`：刑罚的时长，duration 类型，如 `duration({{years:2, months:5}})`。（可缺失）
  - `罚金`：罚款金额，数值类型（单位为元）。（可缺失）

#### 16. **赔偿量**（可缺失）
- **定义**：人物或机构需要赔偿的资产。
- **方向**：`人物 -> 实体资产` 或 `机构 -> 实体资产`。
- **属性**：
  - 无。

#### 17. **赔偿给**（可缺失）
- **定义**：资产赔偿的对象。
- **方向**：`实体资产 -> 人物` 或 `实体资产 -> 机构`。
- **属性**：
  - 无。
"""

````

### 输入到大语言模型生成代码

#### 定义 OpenAI API 密钥

使用支持 OpenAI API 格式的模型，定义好它的Base URL 和 API Key 并指定模型。*需要在 `.streamlit/secrets.toml` 中配置好对应的密钥*

**使用的是 Qwen-VL API，支持 OpenAI API 格式**

````python
import openai
import streamlit as st

# 设置 OpenAI API 密钥（从 Streamlit Secrets 中获取）
# 原使用的是 Qwen-VL API，支持 OpenAI API 格式
# 需要在 `.streamlit/secrets.toml` 中配置好对应的密钥
openai.api_key = st.secrets["OPENAI_API_KEY"]
openai.base_url = st.secrets["OPENAI_BASE_URL"]
model = st.secrets["OPENAI_MODEL"]
````

#### 调用大语言模型生成 Cypher 代码

调用大语言模型生成 Cypher 代码，将生成的 Cypher 代码保存到文件中。

````python
from concurrent.futures import ThreadPoolExecutor, as_completed

# 生成 Cypher 代码
cypher_dir = "out"   # 生成的 Cypher 代码保存目录
os.makedirs(cypher_dir, exist_ok=True)

def generate_cypher(txt_name, text):
    cypher_name = txt_name.removesuffix(".txt") + ".cypher" # 生成的 Cypher 文件名
    if os.path.exists(os.path.join(cypher_dir, cypher_name)): # 如果文件已存在，则跳过
        print(f"文件 {cypher_name} 已存在，跳过。")
        return
    print(f"正在生成 Cypher 代码：{cypher_name}")
    
    prompt = prompt_template.format(text=text)  # 套用模板

    # 调用 OpenAI API 生成 Cypher 代码
    response = openai.chat.completions.create(
        model=model,
        messages=[{"role": "system", "content": prompt}],
        temperature=0.2 # 因为需要生成代码，所以降低温度以减少随机性
    )
    
    # 保存生成的 Cypher 代码
    cypher_code = response.choices[0].message.content
    with open(os.path.join(cypher_dir, cypher_name), "w") as file:
        file.write(cypher_code)
        print(f"已生成 Cypher 代码：{cypher_name}")
````

#### 并发请求 API

由于大语言模型的调用是一个耗时的过程，可以使用多线程并发请求 API，加快生成速度。

**Qwen-VL API 支持并发请求，可以同时请求多个 API**

````python
# 使用线程池并发处理
with ThreadPoolExecutor(max_workers=len(texts)) as executor:
    futures = [executor.submit(generate_cypher, txt_name, text) for txt_name, text in texts.items()]
    for future in as_completed(futures):
        try:
            future.result()
        except Exception as e:
            print(f"生成 Cypher 代码时出错: {e}")
````

## 导入知识图谱数据到 Neo4j 数据库

大语言模型生成的 Cypher 代码可以直接导入到 Neo4j 数据库中，生成知识图谱。

### 连接到 Neo4j 数据库

````python
import neo4j
import streamlit as st
from concurrent.futures import ThreadPoolExecutor, as_completed
from shutil import copyfile, move
import os


# 连接 Neo4j 数据库
uri = st.secrets["NEO4J_URI"]
username = st.secrets["NEO4J_USERNAME"]
password = st.secrets["NEO4J_PASSWORD"]
database = st.secrets["NEO4J_DATABASE"]

driver = neo4j.GraphDatabase.driver(uri, auth=(username, password), database=database)
````

````python
import re

cypher_dir = "out"   # 需要导入到 Neo4j 的 Cypher 代码保存目录（即生成的 Cypher 代码所在目录）

# 执行 Cypher 代码
error_cypher_names = {}
def execute_cypher(cypher_name):
    cypher_path = os.path.join(cypher_dir, cypher_name)
    with open(cypher_path, "r") as file:
        cypher_code = file.read()
        print(f"正在执行 Cypher 代码：{cypher_name}")
        try:
            with driver.session() as session:
                # 把原 txt 的内容也导入到数据库
                text = texts.get(cypher_name.removesuffix(".cypher") + ".txt")
                if text == None:
                    raise Exception("找不到对应的 txt")
                # 查找 Cypher 代码中的 `case*` 变量`
                case_var_names = set(re.findall(r'(?<![a-zA-Z0-9_])(case\d*)(?![a-zA-Z0-9_])', cypher_code))
                for case_var_name in case_var_names:
                    cypher_code = cypher_code + "\n" + f"SET {case_var_name}.content = " + repr(text) # 使用 repr 函数转义字符串
                session.run(cypher_code)
            print(f"已执行 Cypher 代码：{cypher_name}")
        except Exception as e:
            print(f"执行 Cypher 代码时出错: {cypher_name} - {e}")
            os.makedirs("error_cypher", exist_ok=True)
            copyfile(cypher_path, os.path.join("error_cypher", cypher_name))
            error_cypher_names[cypher_name] = e
            
````

### 并发导入到 Neo4j

由于生成的 Cypher 代码量较大，可以使用多线程并发导入到 Neo4j 数据库中，加快导入速度。

**需要注意的是，Neo4j 的事务一致性较差，如果并发导入数据量较大，可能会导致数据不一致，需要根据实际情况调整并发数量。**

````python
# 使用线程池并行处理
with ThreadPoolExecutor(max_workers=1) as executor:
    futures = [executor.submit(execute_cypher, cypher_name) for cypher_name in os.listdir(cypher_dir)]
    for future in as_completed(futures):
        try:
            future.result()
        except Exception as e:
            print(f"执行 Cypher 代码时出错: {e}")
````
